## Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Here are five common plots that can be created using Seaborn library and their uses:

1. Scatter plot: It is used to display the relationship between two numerical variables. It can be helpful in identifying any patterns, trends or outliers in the data.

2. Line plot: It is used to show the trend of a numerical variable over time. It is useful in showing how a variable changes over a certain time period.

3. Histogram: It is used to represent the distribution of a single numerical variable. It can help in identifying the shape, central tendency and spread of the data.

4. Box plot: It is used to display the distribution of a numerical variable across different categories. It is useful in identifying any outliers, skewness or differences between groups.

5. Heatmap: It is used to display the relationship between two categorical variables. It can help in identifying any patterns or trends in the data, and also identify any strong or weak associations between the variables.

## Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fm = sns.load_dataset('fmri')
fm
```

```python
sns.lineplot(x= fm.timepoint , y=fm.signal,data= fm, style='region', hue='event')
```

## Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
# importing pyplot for subploting so that both the charts come in one chart
import matplotlib.pyplot as plt 
titan = sns.load_dataset('titanic')
```

```python
# assigning variable to data series
x = titan['pclass']
y1 = titan['age']
y2 = titan['fare']

# assigning boxes
box1 = [x,y1]
box2= [x,y2]
```

```python
# sizing the graph
fig = plt.figure(figsize=(8,4))
sns.set_style('whitegrid')  #styling the graph

# setting the location for 1st chart
plt.subplot(1,2,1)
lab1= sns.boxplot(box1)  #plotting
lab1.set(title="Box1", xticklabels=['pclass','age'])  #labeling

# setting the location for 2nd chart
plt.subplot(1,2,2)
lab2 = sns.boxplot(box2)  #plotting
lab2.set(xticklabels=['pclass', 'fare'], title='Box2')   #labeling

```

## Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
dia = sns.load_dataset('diamonds')
```

```python
sns.histplot(x=dia.price, data=dia, hue='cut', multiple='stack')
```

## Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iri = sns.load_dataset('iris')
```

```python
sns.pairplot(iri, hue= 'species')
```
